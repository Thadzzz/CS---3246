# **Case Study 1: Iris Flower Classification** 🌸  

### **Background**  
A botanical research institute wants to develop an automated system that classifies different species of **iris flowers** based on their **sepal and petal measurements**.  The dataset consists of **150 samples**, labeled as **Setosa, Versicolor, or Virginica**.  

### **Problem Statement**  
Can we use **sepal and petal dimensions** to correctly classify the **species of an iris flower**?  

### **Task Description**  

#### **1. Data Exploration**  
- Load the dataset and display the first few rows.  
- Identify any missing or inconsistent values.  

#### **2. Data Cleaning**  
- Check for missing values and handle them appropriately.  
- Convert categorical species labels into a format suitable for analysis.  

#### **3. Basic Data Analysis**  
- Find the average sepal and petal dimensions for each species.  
- Identify correlations between different flower measurements.  

#### **4. Visualization**  
- Create simple visualizations (e.g., histograms, scatter plots) to understand data distribution.  

#### **5. Insights & Interpretation**  
- Summarize key findings, such as which features best distinguish flower species.  

```python
import pandas as pd

# Loads the dataset
iris = pd.read_csv('iris_dataset.csv')

# For showing the first few rows
display(iris.head())

# This checks for missing/inconsistent values
display(iris.info())

# Looks for missing values (adds up if so)
display(iris.isnull().sum())

# Makes the species column into a categorical form (easier to understand)
iris['species'] = iris['species'].astype('category')

# For finding the average of sepal and petal dimensions
average_dimensions = iris.groupby('species').mean()
display(average_dimensions)

# For corellation between flowe measurements
correlation_matrix = iris.select_dtypes(include=[float, int]).corr()
display(correlation_matrix)

import matplotlib.pyplot as plt
import seaborn as sns

# Visualization via histogram
iris.hist(edgecolor='black', figsize=(12, 8))
plt.show()

# Scatter plot for sepal length VS sepal width colored by species
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

**1B Insights & Interpretation:** Personally I find the histogram to be the most convenient feature in regards to displaying the differemt height/width measurements of each flower species. But that's not to say the other ones are useless. I just think it's the best one to use in general as it is easier to identify the differences between everything.

# **Case Study 2: Netflix Content Analysis** 🎬  

## **Background**  
Netflix is a leading streaming platform with a vast collection of movies and TV shows. The company wants to analyze its **content library** to understand trends in **genres, release years, and regional distribution**.  

## **Problem Statement**  
How can we use **Netflix’s dataset** to gain insights into content distribution, popular genres, and release trends over time?  

## **Task Description**  

### **1. Data Exploration**  
- Load the dataset and inspect its structure.  
- Identify key columns such as title, genre, release year, and country.  

### **2. Data Cleaning**  
- Check for missing or incorrect values in key columns.  
- Remove duplicates and format the date-related data properly.  

### **3. Basic Data Analysis**  
- Count the number of movies vs. TV shows.  
- Identify the most common genres and countries producing content.  
- Analyze the number of releases per year to observe trends.  

### **4. Insights & Interpretation**  
- Summarize key findings, such as trends in Netflix's content production over time.  


```python
import pandas as pd

# Loads the dataset
netflix_data = pd.read_csv('netflix_dataset.csv')

# Shows the first few rows
display(netflix_data.head())

# Displays the general information and data types
display(netflix_data.info())

# Checks for missing values
display(netflix_data.isnull().sum())

# Removes duplicates
netflix_data.drop_duplicates(inplace=True)

# Converts 'date_added' to datetime
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'], errors='coerce')

# Makes sure that 'release_year' is an integer
netflix_data['release_year'] = pd.to_numeric(netflix_data['release_year'], errors='coerce')

# Drop rows with missing 'release_year' (if needed)
netflix_data.dropna(subset=['release_year'], inplace=True)

# Counts the number of movies compared to TV shows
display(netflix_data['type'].value_counts())

# Using a separate Series to handle genre exploration after splitting and exploding
listed_in_series = netflix_data['listed_in'].str.split(', ').explode()
# Now count the most common genres using the newly created Series
display(listed_in_series.value_counts().head(10))

# Lists the total amount of shows made for each country
display(netflix_data['country'].value_counts().head(10))

# Shows how much shows were released for a particular year
display(netflix_data['release_year'].value_counts().sort_index())
```
